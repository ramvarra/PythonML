```python
import matplotlib.pyplot as plt
%matplotlib inline
import os
from PIL import Image

import logging
```

```python
src_dir = r'N:\FAMILY\MEDIA\2017\2017-06-09 Kishore Padmini Visit'
dst_dir = r'C:\TEMP\Thumbnails'
```

```python

def create_thumbs(f):
    print('Converting {} to thumbnail'.format(f))
    src_img = Image.open(os.path.join(src_dir, f))
    src_img.thumbnail((64, 64), Image.ANTIALIAS)
    src_img.save(os.path.join(dst_dir, f))
    print('{} completed.'.format(f))
    
for f in os.listdir(src_dir):
    create_thumbs(f)
```
