```python
import numpy as np
import pandas as pd
```

```python
def conv(s):
    if s.startswith('Same'):
        return 0
    if s.startswith('Below'):
        return -1
    if s.startswith('Above'):
        return 1
    return 0

def col_transform(s):
    return '_'.join(s.split()[:-2])

df = pd.read_csv(r'data\Hospital General Information.csv', encoding='latin-1')
input_cols = [c for c in df.columns if c.endswith('comparison')] 
target_col = ['Hospital overall rating']
df_input = df[input_cols].applymap(conv)
df_input.columns = [col_transform(s) for s in df_input.columns]
df_target = df[target_col]
df_target.columns = ['Rating']
df_target = df_target[df_target.Rating != 'Not Available']
df_input = df_input.loc[df_target.index, :]
```
