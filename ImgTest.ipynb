```python
dir_name = r'N:\FAMILY\MEDIA\2017\2017-06-09 Kishore Padmini Visit'

img1 = Image.open(os.path.join(dir_name, 'DSC_2490.jpg'))
img2 = Image.open(os.path.join(dir_name, 'DSC_2455.jpg'))
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,10))
axs[0].imshow(img1)
axs[1].imshow(img2)
```
